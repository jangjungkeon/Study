```python
student_score = raw_data.loc[:, ["sex", "G1", "G2", "G3"]]
student_score.groupby("sex").describe()
```

```python
import numpy as np
from scipy import stats
```

```python
female_math_score = student_score[student_score.sex == "F"].drop("sex", axis=1)
male_math_score = student_score[student_score.sex == "M"].drop("sex", axis=1)

# ì •ê·œì„±
# female_shapiro = stats.shapiro(female_math_score)
# male_shapiro = stats.shapiro(male_math_score)
# print(f"female_shapiro : {female_shapiro}, male_shapiro : {male_shapiro}")

# ë“±ë¶„ì‚°ì„±
female_bartlett = stats.bartlett(female_math_score.iloc[:, 0], male_math_score.iloc[:, 0])
# male_bartlett = stats.bartlett(male_math_score.columns)
#print(f"female_bartlett : {female_bartlett}, male_bartlett : {male_bartlett}")

for i in range(female_math_score.shape[1]):
    female_shapiro = stats.shapiro(female_math_score.iloc[:, i])
    male_shapiro = stats.shapiro(male_math_score.iloc[:, i])
    bartlett = stats.bartlett(female_math_score.iloc[:, i], male_math_score.iloc[:, i])
    print(f"""{i+1}ë²ˆì§¸ female_shapiro : {female_shapiro.pvalue}, male_shapiro : {male_shapiro.pvalue}, 
           bartlett : {bartlett.pvalue}""")

result = stats.ttest_ind(male_math_score, female_math_score, alternative='less')
print(result)
result2 = stats.ttest_rel(raw_data.G1, raw_data.G3, alternative='two-sided')
print(result2)
```
